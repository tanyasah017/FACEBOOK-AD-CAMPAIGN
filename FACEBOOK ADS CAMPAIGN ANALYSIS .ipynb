# REGRESSION ANALYSIS

## Assigning independant variables(X) and dependant variable(y)

```python
X=df[['total_conversion','approved_conversion']]
y=df[['spent']]
```

# Importing train test split

```python
from sklearn.model_selection import train_test_split
```

## Setting test size to 20%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Importing linear regression

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

# Fitting of the data and predicting the values

```python
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

## Finding out the R-squared value

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,pred)
r2
```

The R-squared value is 0.56
